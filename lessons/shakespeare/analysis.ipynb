# Quantified Shakespeare

The purpose of this notebook is to explore character/letter frequency in Shakespeare's works.

Steps:

1. Read data into list of lines
2. Count characters in a dictionary
3. Print results

```python
lines: list[str] = []
file_handle = open("shakespeare.txt", "r", encoding="utf8")
lines = file_handle.readlines()
file_handle.close()
print(f"{len(lines)} lines of Shakespeare read")
lines[5]
```

Now that we have all lines of text in a dictionary named `lines`,
let's count the characters Shakespeare used!

```python
frequencies: dict[str, int] = {}

#1. Iterate over every line in lines
# 2. Iterate over every character in line
#  3. Incrase the frequency count of the current character by 1
for line in lines:
    for character in line:
        if character in frequencies:
            frequencies[character] += 1
        else:
            frequencies[character] = 1

frequencies 
```

### Let's avoid counting non-alphabetical characters

```python
for line in lines:
    for character in line:
        if character.isalpha():        
            character = character.lower()
            if character in frequencies:
                frequencies[character] += 1
            else:
                frequencies[character] = 1

frequencies
```
